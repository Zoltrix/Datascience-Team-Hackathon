```python
def test_func(x):
    return [item for sublist in x['ingredients'] for item in sublist]
```

```python
cusine_ingridents = df.groupby('cuisine').apply(test_func)
```

```python
cusine_ingridents
```

```python
cusine_ingridents = pd.DataFrame(cusine_ingridents).reset_index()
```

```python
cusine_ingridents.columns = ['cuisine','total_ingredients']
```

```python
cusine_ingridents
```

```python
cusine_ingridents['unique_ingredients'] = cusine_ingridents['total_ingredients'].apply(lambda x: list(set(x)))
```

```python
cusine_ingridents['no_unique_ingredients'] = cusine_ingridents['unique_ingredients'].apply(lambda x: len(x))
```
